```python
x2 = np.arange(0, 4*np.pi, 0.05)
ycos = np.cos(x2)
ysin = np.sin(x2)

fig, ax = plt.subplots()
ax.plot(x2, ycos, 'b-', label = 'cos(x)')
ax.plot(x2, ysin, 'r--', label = 'sin(x)')
ax.set_title('Triginometric functions')
ax.set_xlabel('0 to 4pi')
ax.set_ylabel('cos (x) and sin (x)')
ax.legend()
```

```python
#Bar plot

people = ["Student A", "Student B"]
StudentA = np.array([90,50,80,40])
StudentB = np.array([75,45,60,95])
p = np.arange(len(StudentA))
q = np.arange(len(StudentB))

fig, ax = plt.subplots(1,2)
ax[0].bar(p, StudentA, width = 0.5)
ax[1].bar(p, StudentB, width = 0.5)

fig.tight_layout()

for i in range(len(ax)):
    ax[i].set_ylim(0,100)
    ax[i].set_title('People')
    ax[i].set_xlabel('Exercises')
    ax[i].set_ylabel('mark(%)')
    ax[i].set_xticks([0,1,2,3])
    ax[i].set_xticklabels(['ex1','ex2','ex3','ex4'])




```

```python
fig, ax =plt.subplots()
width = 0.3
s1bars = ax.bar(p - width/2, StudentA, width)
s2bars = ax.bar(p + width/2, StudentB, width)
```

```python
##Bar Plot (grouped)
fig, ax = plt.subplots()
width = 0.3
s1bars = ax.bar(p - width/2, StudentA, width, label='Student A')
s2bars = ax.bar(p + width/2, StudentB, width, label='Student B')
ax.set_ylim([0, 100])
ax.set_title('Student Scores')
ax.set_xlabel("exercises")
ax.set_ylabel("mark (%)")
ax.set_xticks([0,1,2,3])
ax.set_xticklabels(["ex1","ex2","ex3","ex4"])
ax.legend()
for i in range(len(StudentA)):
    s1bars[i].set_linewidth(3.5)
    s2bars[i].set_linewidth(3.5)
    if StudentA[i] < 50:
        s1bars[i].set_edgecolor('red')
    if StudentB[i] < 50:
        s2bars[i].set_edgecolor('red')
```

```python
#SHAPES

from matplotlib.patches import Polygon
```

```python
fig, ax = plt.subplots()
ax.set_ylim([0,20])
ax.set_xlim([-10,10])
ax.axis('scaled')

start_x = -2.5
start_y = 2
width = 5
height = 6
facecolor = 'gray'

##create a rectangle
rec=plt.Rectangle((start_x,start_y), width, height, fc= facecolor)
ax.add_patch(rec)

## add patch
rec1=plt.Rectangle((-0.5, 2), 1,2, fc= 'brown')
ax.add_patch(rec1)

## add a circle

circle1 = plt.Circle([-1, 6], 0.5, fc = 'yellow', edgecolor = 'black', lw = 1)
ax.add_patch(circle1)

circle2 = plt.Circle([1, 6], 0.5, fc = 'yellow', edgecolor = 'black', lw = 1)
ax.add_patch(circle2)

## add triangle

start_xy = np.array([-2.5, 8])
end_xy = np.array([2.5,8])
top_xy = np.array([0,10])
tri = Polygon(np.vstack([start_xy, end_xy, top_xy]), color= 'red')
ax.add_patch(tri)
```
