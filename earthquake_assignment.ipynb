```python
## SCATTER PLOT

import matplotlib.pyplot as plt

# Latitude and Longitude data 
latitude = earthquake_data['latitude']  # List of latitude values
longitude = earthquake_data['longitude']  # List of longitude values

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(longitude, latitude, color='red', marker='o', s=10)  # s is the size of markers
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Earthquake Epicenters Scatter Plot')
plt.grid(True)
plt.show()
```

```python
# Select the last 30 events of data
head_30days = earthquake_data.head(30)
```

```python
## HISTOGRAM of 'MAG' (magnitude) for the last 30 events
import numpy as np

plt.figure(figsize=(9, 6))
plt.hist(head_30days['mag'], bins=8, alpha=0.9, label='Last 30 events', edgecolor='black')
plt.title('Histogram of Earthquake Magnitudes for the Last 30 events')
plt.xlabel('Magnitude')
plt.xticks(np.arange(0,5, 0.5))
plt.ylabel('Frequency')
plt.legend()

plt.show()
```

```python
# Count the number of occurrences for each magType
magType_counts = earthquake_data['magType'].value_counts()
magType_counts = magType_counts.reset_index()
magType_counts
```

```python
## PIE CHART of Magnitude Types

fig, ax = plt.subplots(figsize=(10, 10))
magType_counts.plot(kind = 'pie', y='count', labels = magType_counts['magType'], autopct =' %1.2f', ax=ax)
plt.title('Earthquake magType Distribution')

plt.show()
```

```python
import calendar
```

```python
# Extract the year and month from the time field
earthquake_data['Year'] = earthquake_data['time'].apply(lambda x: int(x[:4]))
earthquake_data['Month'] = earthquake_data['time'].apply(lambda x: calendar.month_name[int(x[5:7])])
earthquake_data['Day'] = earthquake_data['time'].apply(lambda x: calendar.day_name[int(x[7:9])])
earthquake_data
```

```python
# Group the data by month and calculate the median magnitude
monthly_magnitudes = earthquake_data.groupby('Month')['mag'].median()
monthly_magnitudes
```

```python
# Create a histogram with the month names on the x-axis
plt.figure(figsize=(7, 4))
plt.bar(monthly_magnitudes.index, monthly_magnitudes.values, width = 0.5)
plt.title('Histogram of Earthquake Magnitudes by Month')
plt.xlabel('Month')
plt.ylabel('Median Magnitude')
plt.show()
```
