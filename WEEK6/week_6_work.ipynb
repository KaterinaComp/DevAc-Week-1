```python
import pandas as pd
import numpy as np

# Sample data
data = {
    'food_id': [24, 70, 192, 400],
    'food_name': ['shish kebab', 'rice', np.nan, 'syrup'],
    'date_added': ['2019-08-08 00:00:00', '2018-03-23 00:00:00', '2019-01-20 00:00:00', np.nan],
    'date_modified': [np.nan, '2019-03-10 00:00:00', '2019-09-03 00:00:00', '2019-09-04 00:00:00'],
    'total_sales & online_sales': ['NaN', '0 (0)', '2596 (2791)', '808 (2502)'],
    'approved': [1.0, np.nan, 1.0, 1.0]
}

# Create DataFrame
df = pd.DataFrame(data)

# 1. Correcting NaN values in 'food_name' column
df['food_name'] = df['food_name'].fillna('Unknown')

# 2. Correcting NaN values and converting date strings to datetime objects in 'date_added' and 'date_modified' columns
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['date_modified'] = pd.to_datetime(df['date_modified'], errors='coerce')

# 3. Correcting 'total_sales & online_sales' column by splitting the values and converting them to integers
def process_sales(sales_str):
    if sales_str == 'NaN':
        return np.nan
    else:
        return int(sales_str.split()[0])

df['total_sales'] = df['total_sales & online_sales'].apply(process_sales)

# 4. Correcting 'approved' column by filling NaN values with 0
df['approved'] = df['approved'].fillna(0)

# Display the corrected DataFrame
print(df)
```

```python
## read data
data = pd.read_csv('fruit_types.csv')
data.head()
```

```python
data.info()
```

```python
## missing values
data.isnull().sum()
```

```python
## Label encoder
encoder = LabelEncoder()

## encode all variables in dataset
fruit_name_encoded = encoder.fit_transform(data.iloc[:,0])
fruit_subtype_encoded = encoder.fit_transform(data.iloc[:,1])
mass_encoded = encoder.fit_transform(data.iloc[:,2])
width_encoded = encoder.fit_transform(data.iloc[:,3])
height_encoded = encoder.fit_transform(data.iloc[:,4])
color_score_encoded = encoder.fit_transform(data.iloc[:,5])
```

```python
##combine the encoded data
data = list(zip(fruit_name_encoded, fruit_subtype_encoded, mass_encoded, width_encoded,height_encoded, color_score_encoded))
```

```python

##to put data into a dataframe
data = pd.DataFrame(data,columns=['fruit_name', 'fruit_subtype','mass','width','height','color_score'])
data.head()
```

```python
## split our data
data_to_use = data.iloc[:,1:6]
data_to_target = data.iloc[:,0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_to_use,data_to_target, test_size=0.3, random_state=32)
```

```python
##buoild KNN model

knn_model =KNeighborsClassifier(n_neighbors=1)
knn_model.fit(X_train,y_train)
```

```python
## make prediction
y_pred_knn = knn_model.predict(X_test)
```

```python
print('Accuracy', metrics.accuracy_score(y_test,y_pred_knn))
```

```python
from math import sqrt
```

```python
## root mean square error
print('RMSE', sqrt(metrics.mean_squared_error(y_test,y_pred_knn)))
```
